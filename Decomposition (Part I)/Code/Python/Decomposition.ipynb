```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
```

```python
data = pd.read_csv("Monthly-anti-diabetic.csv", index_col="Month")
# data["Date"] = pd.to_datetime(data["Date"])
# data.set_index(data["Date"], inplace=True) 
# data.drop("Date", inplace=True, axis=1)
```

```python
# data
plt.plot(data)

plt.xlabel("Month")
plt.show()
```

### Python Additive Decompose

```python
sea_dec = seasonal_decompose(data["Monthly Demand in Millions"], model='additive', freq=12)
fig_size = plt.rcParams["figure.figsize"]
fig_size[0], fig_size[1] = 16, 12
plt.rcParams["figure.figsize"] = fig_size
sea_dec.plot()
plt.xticks(rotation='vertical')
plt.show()
```

```python
sea_dec.resid
```

## Python Multiplicative Decompose

```python
sea_dec = seasonal_decompose(data["Monthly Demand in Millions"], model='multiplicative', freq=12)
fig_size = plt.rcParams["figure.figsize"]
fig_size[0], fig_size[1] = 16, 12
plt.rcParams["figure.figsize"] = fig_size
sea_dec.plot()
plt.xticks(rotation='vertical')
plt.show()
```

```python
sea_dec.resid
```

### Python Decomposition by hand

```python
data["Monthly Moving Average 2 * 12"] = data["Monthly Demand in Millions"].rolling(window=12, center=True).mean().rolling(window=2, center=True).mean()
```

```python
data
```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0], fig_size[1] = 16, 3
plt.rcParams["figure.figsize"] = fig_size
plt.plot(data["Monthly Moving Average 2 * 12"])
plt.xticks(np.arange(0, 220, step=5), rotation='vertical')
plt.show()
```

```python
data["detrended"] = data["Monthly Demand in Millions"] - data["Monthly Moving Average 12"]
```
